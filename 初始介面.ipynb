```python
#新增
import tkinter as tk
from tkinter import ttk
import os
from PIL import ImageTk, Image
import base64
from io import BytesIO
import sys
sys.path.append(r"C:\Users\c0910\Downloads") #匯入解魔方步驟的py檔案路徑
import threading



class CubeSolverApp:
    def __init__(self, root):
        self.root = root
        self.root.title("魔方大師")
        self.root.geometry("800x550")
        self.root.resizable(width=True, height=True) #可縮放視窗
        
        self.style = ttk.Style()
        self.style.configure('TButton',padding=10, font=('C:\Windows\Fonts\kaiu.ttf', 15)) #標楷體/大小15
        self.img_path = r'C:\Users\c0910\OneDrive\桌面\新增資料夾\123.jpg'
        self.load_and_display_image()

        self.label = ttk.Label(root, text="請準備好掃描魔方")
        self.label.pack(pady=10)
        #按鈕
        self.solve_button = ttk.Button(root, text="開始", command=self.solve_cube, style='TButton')
        self.solve_button.pack(pady=10)
        self.instructions_button = ttk.Button(root, text="顯示操作說明", command=self.show_instructions) 
        self.instructions_button.pack(pady=10)
        self.quit_button = ttk.Button(root, text="退出", command=self.root.destroy)
        self.quit_button.pack(pady=10)
                
    def load_and_display_image(self):
        # 轉base64再解碼圖像，直接用路徑顯示圖像一直出bug
        with open(self.img_path, 'rb') as img_file:
            img_data = base64.b64encode(img_file.read()).decode('utf-8')

        img_bytes = base64.b64decode(img_data)
        img = Image.open(BytesIO(img_bytes))

        self.photo_image = ImageTk.PhotoImage(img)
        self.canvas = tk.Canvas(self.root, width=img.width, height=img.height, bd=0, cursor='circle')
        self.canvas.create_image(0, 0, anchor=tk.NW, image=self.photo_image)
        self.canvas.pack()
        
            
    def show_instructions(self):
        #創建新視窗
         instructions_window = tk.Toplevel(self.root)
         instructions_window.geometry("600x400")
         instructions_window.resizable(width=True, height=True)
         instructions_window.title("操作說明")

        #新視窗操作說明
         instructions_text = """
         1. 
         準備好要掃描的2*2魔方。
         2. 
         點擊「開始」按鈕，開始對著鏡頭掃描魔方。
         3. 
         跟隨箭頭指示以面對自己的那面魔方轉動魔方來掃描。
         4. 
         開始掃描後以空白鍵進行下一步，並以Esc鍵做退出。
         """
        #內容可持續增加

         instructions_label = ttk.Label(instructions_window,
         text=instructions_text,
         wraplength=700, #一行字最大寬
         justify="left",
         font=('C:\Windows\Fonts\kaiu.ttf', 15)) #字體設定是標楷體，若無法運行可改為'Courier New'
         instructions_label.pack(padx=10, pady=10, anchor='w')

    def solve_cube(self):
        self.label.config(text="正在解魔方，請稍候...")
        threading.Thread(target=self.run_camerapy).start() #讓初始介面在掃描開始後仍可使用

    def run_camerapy(self):
        import camerapy #在這裡才匯入是想避免它早於初始介面出現，應該有其他方法做匯入
        camerapy.run()  #運行camerapy
    



if __name__ == "__main__":
    root = tk.Tk()
    app = CubeSolverApp(root)
    root.mainloop()
```
